```python
import numpy as np
from dataset import *
import matplotlib.pyplot as plt
```

```python
dataset = Dataset("data/train.csv", 'train', False, True)
dataset.showlabel()
```

```python
data = []
subscribte = []
for j in range(0, 22500):
    data_j, subscribte_j = dataset.__getitem__(j)
    data.append(data_j)
    subscribte.append(subscribte_j)
data = np.array(data)
subscribte = np.array(subscribte)
fields = dataset.get_field()
```

```python
plt.figure(1, figsize=(20, 20))
font1 = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 18,
        }
for i in range(20):
    plt.subplot(5, 4 , i + 1)
    if i == 20:
        plt.hist(subscribte[:,0])
    else:
        plt.hist(data[:,i])
#     plt.axis("off")
    plt.title("%s" % fields[i+1], fontdict=font1)
plt.savefig('statics/Data_distribution.png', pad_inches=0) #, bbox_inches='tight')
plt.show()
plt.close('all')
```

```python
import paddle

x = np.linspace(-5, 5, 100)
x = paddle.to_tensor(x)
relu = paddle.nn.ReLU()
out = relu(x)
plt.plot(x.numpy(), out.numpy())
plt.title('ReLU', fontdict={"family": "Times New Roman", "size": 18})
plt.xlabel('x', fontdict={"family": "Times New Roman", "size": 18})
plt.ylabel('y', fontdict={"family": "Times New Roman", "size": 18})
plt.xticks(fontname="Times New Roman", fontsize=18)
plt.yticks(fontname="Times New Roman", fontsize=18)
```
